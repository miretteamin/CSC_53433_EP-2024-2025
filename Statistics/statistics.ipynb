```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
with open('animals.csv', 'r') as file:
    reader = csv.reader(file)
    row = next(reader)
    animals = np.array([int(value) for value in row])

with open('grass.csv', 'r') as file:
    reader = csv.reader(file)
    row = next(reader)
    grass = np.array([int(value) for value in row])

with open('max_generation.csv', 'r') as file:
    reader = csv.reader(file)
    row = next(reader)
    maxGeneration = np.array([int(value) for value in row])

with open('max_lifetime.csv', 'r') as file:
    reader = csv.reader(file)
    row = next(reader)
    maxLifetime = np.array([int(value) for value in row])

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes[0,0].plot(animals);
axes[0,0].set_title("Number of Animals over time", fontsize=16);
axes[0,0].set_xlabel("Time", fontsize=12);
axes[0,0].set_ylabel("Number of Animals", fontsize=12);

axes[0,1].plot(grass);
axes[0,1].set_title("Number of Grass over time", fontsize=16);
axes[0,1].set_xlabel("Time", fontsize=12);
axes[0,1].set_ylabel("Number of Grass", fontsize=12);

axes[1,0].plot(maxGeneration);
axes[1,0].set_title("Max Generation over time", fontsize=16);
axes[1,0].set_xlabel("Time", fontsize=12);
axes[1,0].set_ylabel("Max Generation", fontsize=12);

axes[1,1].plot(maxLifetime);
axes[1,1].set_title("Max Lifetime over time", fontsize=16);
axes[1,1].set_xlabel("Time", fontsize=12);
axes[1,1].set_ylabel("Max Lifetime", fontsize=12);
```
